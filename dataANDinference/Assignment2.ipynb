Que.4.

The World Bank Indicators provide variables called “Fertility rate,
total (births per woman)” and “GDP per capita (current US$)”.
Make a scatter plot of Fertility rate versus GDP per capita for all
countries in 2010. Produce cumulative distribution functions for the
fertility rate variable using data from 1990 and 2010 respectively.
Use vertical lines to indicate the mean and median. Use a legend to
explain which is the mean and which is the median. Have fertility
rates changed over this twenty-year period?

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

fertility  = pd.read_csv('fertility.csv')['2010']
gdpusd = pd.read_csv('gdpusd.csv')['2010']

gdpusd = gdpusd.apply(lambda x: (x - gdpusd.min())/(gdpusd.max() - gdpusd.min()))
gdpusd = gdpusd.clip(lower=0.01, upper=0.99)

plt.scatter(fertility, gdpusd, label='gdp per capita')

fertility1990  = pd.read_csv('fertility.csv')['1990']
#to calc cdf, you need to sort

fertility1990_sorted = fertility1990.sort_values()
fertility2010_sorted = fertility.sort_values()

fertility1990_sorted[len(fertility1990_sorted)] = fertility1990_sorted.iloc[-1]
fertility1990_cum_dis = np.linspace(0.,1.,len(fertility1990_sorted))
fertility1990_cdf = pd.Series(fertility1990_cum_dis, index=fertility1990_sorted)

plt.plot(fertility1990_cdf, label='1990 Fertility CDF')

meanfertility1990 = np.nanmean(fertility1990)

plt.axvline(x = meanfertility1990, color = 'r', label='1990 Fertility Mean')

medianfertility1990 = np.nanmedian(fertility1990)

plt.axvline(x = medianfertility1990, color = 'c', label='1990 Fertility Median')

fertility2010_sorted[len(fertility2010_sorted)] = fertility2010_sorted.iloc[-1]
fertility2010_cum_dis = np.linspace(0.,1.,len(fertility2010_sorted))
fertility2010_cdf = pd.Series(fertility2010_cum_dis, index=fertility2010_sorted)

plt.plot(fertility2010_cdf, label='2010 Fertility CDF')
meanfertility2010 = np.nanmean(fertility)

plt.axvline(x = meanfertility2010, color = 'm', label='2010 Fertility Mean')

medianfertility2010 = np.nanmedian(fertility)

plt.axvline(x = medianfertility2010, color = 'y', label='2010 Fertility Median')



plt.ylabel('GDP or Probability')
plt.xlabel('Fertility')
plt.legend()
plt.show()
```

![title](Q4grp1.png)

The cumulative distribution function (cdf)for this question shows the probability that the fertility takes a value less than or equal to a value x. After reading the data the cummulative distribution was simply gotten by sorting the data, getting evenly spaced normalized numbers over 0 - 1, and this was plotted against the sorted data. 

The fertiliy rate has indeed decreased over this 20 year period. With the mean fertility rate shrinking to
2.89 in 2010 from 3.96 in 1990. Also the median fertility has also decreased from 3.60 to 2.39 within those 20 years.


Que. 5.

Download the latest data for the “Happy Planet Index” from
http://www.happyplanetindex.org and “Corruption Perceptions Index” from http://www.transparency.org/cpi2014/in_detail . Both datasets are available as excel spread-sheets.
Find matching countries for both indices and make a carefully labelled scatter plot
to demonstrate the relationship using ranks in both cases. Are there
any countries that stand out as being unusual? 

```python
import pandas as pd
import numpy as np
from numpy.random import rand
import matplotlib.pyplot as plt
import matplotlib.cm as cm


hpi = pd.read_csv('hpi.csv').dropna()
cpi = pd.read_csv('CPI.csv').dropna()

hpi_countries = hpi['Country']

index = 0

Ranks = pd.DataFrame({'Country': ["" for x in range(len(cpi))], 'CPI': [0 for x in range(len(cpi))],
                      'HPI': [0 for x in range(len(cpi))]})

for country_name in cpi.Country:
   
    hpiindexlist = (hpi_countries[hpi_countries == country_name]).index.tolist()
    hpiindex = -1
    
    if len(hpiindexlist) > 0:
    
        hpiindex = hpiindexlist[0]
        Ranks.CPI[index] = index +1
        Ranks.HPI[index] = hpiindex + 1
        Ranks.Country[index] = country_name
        
    index = index + 1

CPI = Ranks.CPI[Ranks.CPI != 0]
HPI = Ranks.HPI[Ranks.HPI != 0]
Country = Ranks.Country[Ranks.Country != ""]

color=iter(cm.rainbow(np.linspace(0,1,len(CPI))))

plt.xlabel('HPI')
plt.ylabel('CPI')


for hpp, cpp,ct, in zip(HPI, CPI, Country):
    plt.scatter(hpp, cpp, color=next(color))
    plt.annotate(
        ct,  xy = (hpp, cpp), xytext = (-20, 20), textcoords = 'offset points', ha = 'right', va = 'bottom',
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc2,rad=0'))
    
    
plt.show()
```
